```python
plt.figure(figsize=(10, 5))
plt.axes()
plt.hist([1,2,1,2,3,4])
plt.title('ㅗ'); # 제목
# plt.xticks([1,2,3]); # x축 
# plt.yticks([1,3,5,7]); # y축
plt.xlim([1,2]) # x범위 지정
plt.ylim([1,2]) # y범위 지정
plt.xlabel('xTicks Name') # x축 이름
plt.ylabel('yTicks Name') # y축 이름
plt.grid(True) # ticks기준으로 모눈 그림
```
