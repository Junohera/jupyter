```python
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
```

```python
tips
```

```python
# x축 : day y축 : time 가장 작은팀
```

```python
pd.pivot_table(tips, index='day', columns='time',values='tip', aggfunc='min')
```

```python
tips.pivot_table(index='day', columns=['sex', 'time'], values='tip', aggfunc='min')
```

```python
pd.crosstab(index=tips['day'], columns=tips['time'], values=tips.tip, aggfunc='mean')
```

```python
tips
```

```python
tips.values # numpy 포맷으로 변경
```

```python
# 컬럼추출방법 3가지

# 1. series
tips['tip'] 

# 2. dataFrame
tips[['tip']] 

# 3. 프로퍼티로접근
tips.tip 
```
