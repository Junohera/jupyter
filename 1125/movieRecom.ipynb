```python
def recom(user_id) :
    data[(data.user_id==user_id) & (data.rating==5)]['item_id'] # 3. 913번이 봤던 영화 추출(평점5) & 25번이 안본영화(차집합)
    data[data.user_id==25]['item_id'] # 3. 913번이 봤던 영화 추출(평점5) & 25번이 안본영화(차집합)
    t = set(data[(data.user_id==user_id) & (data.rating==5)]['item_id']) - set(data[data.user_id==25]['item_id'])
    return item.loc[t][1] # 4. 3번의 영화제목만 추출
```
