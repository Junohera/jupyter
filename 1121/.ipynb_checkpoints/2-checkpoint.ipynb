```python
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
```

```python
tips
```

```python
tips['size'] # series
```

```python
tips[['size']] # dataFrame
```

```python
tips.day # series
```

```python
tips
```

```python
tips[(tips['sex'] == 'Female') & (tips['smoker'] == 'Yes') & (tips['size'] == 2)] # 여성이면서 담배피는사람들!!
```

```python
tips[(tips['sex'] == 'Female') & (tips['smoker'] == 'Yes') & (tips['size'] == 2)].mean() # 여성이면서 담배피는사람들의 모든 평균!! ()
```

```python
tips[(tips['sex'] == 'Male') & (tips['smoker'] == 'Yes')].max() # 남자면서 담배피는 사람들중에 최대팁알기!
```

```python
tips.rename({'size':'size_'}, axis=1, inplace=True) # size를 size_로 바꾸기!! + 원본수정
```

```python
tips
```

```python
tips.groupby('sex').mean() # 성별로 그룹지어서 평균내기
```

```python
tips.groupby('sex').max() # 성별로 그룹지어서 최대값 구하기
```

```python
tips.groupby('sex').min() # 성별로 그룹지어서 최초값 구하기
```

```python
tips.groupby('sex').mean()['tip'] # 성별로 그룹지어서 평균팁만 출력하고싶다!!
```

```python
tips.groupby('sex').mean()[['tip']] # 성별로 그룹지어서 평균팁만 출력하고싶다!! + dataFrame
```

```python
tips.groupby('sex').mean()[['tip']].plot.bar()
```

```python
tips.groupby('sex').mean()[['tip', 'total_bill']].plot.bar()
```

```python
tips.groupby('sex').mean()[['tip', 'total_bill']].plot.bar(stacked=True)
```

```python
tips.groupby(['sex', 'smoker']).mean()
```

```python
tips.groupby(['sex', 'smoker']).mean().plot.bar()
```

```python
# 목요일날 식사한사람들의 남녀팁의 평균
```

```python
tips[tips['day'] == 'Thur'].groupby('sex').mean()
```

```python
tips[tips['day'] == 'Thur'].groupby('sex').mean().plot.bar()
```

```python
tips[tips['day'] == 'Thur'].groupby('sex').mean()[['tip']].plot.bar()
```

```python
tips[tips['day'] == 'Sun'].groupby('sex').mean()[['tip']].plot.bar()
```

```python
tips[tips['day'] == 'Sun'].groupby('sex').max()[['tip']].plot.bar()
```
