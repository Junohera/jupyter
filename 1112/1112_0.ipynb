```python
class A:
    a = 1
    
    def __init__(self, a):
        self.x = 1
```

```python
class B(A):
    b = 1
    a = 4
    def __init__(self, a):
        super().__init__(a)
        self.a = a
        print('INIT B')
    
    def bb(self):
        self.b = 3
        print('BB')
```

```python
b = B(5)
```
