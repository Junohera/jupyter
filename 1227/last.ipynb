```python
from tensorflow.keras.datasets import mnist
```

```python
# hold out - 
(X_train, y_train), (X_test, y_test) = mnist.load_data()
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(iris.iloc[:,:-1], iris.target)
```

```python
# sklearn datasets, model_selection, metrics, preprocessing
```

```python
from sklearn.neural_network import MLPClassifier
mlp = MLPClassifier()
mlp.fit(X_train, y_train)
mlp.score(X_test, y_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix, classification_report

print(confusion_matrix(y_test, mlp.predict(X_test)))
print(classification_report(y_test, mlp.predict(X_test)))
knn = KNeighborsClassifier()
knn.fit(X_train, y_train)
knn.score(X_test, y_test)
```

```python
from sklearn.model_selection import cross_val_score
```

```python
result = cross_val_score(MLPClassifier(), iris.iloc[:, :-1], iris.iloc[:, -1], cv=10)

print(result)
print(result.mean())

result = cross_val_score(KNeighborsClassifier(), iris.iloc[:, :-1], iris.iloc[:, -1], cv=10)
print(result)
print(result.mean())
```

```python
result.mean()
```

```python
from sklearn.pipeline import Pipeline
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
pipe = Pipeline([('std', StandardScaler()), ('est', KNeighborsClassifier())])
```

```python
vars(pipe)
```

```python
# 가장 좋은 모델 + 하이퍼파라미터찾기
from sklearn.model_selection import GridSearchCV

grid = GridSearchCV(pipe, {'est':[KNeighborsClassifier()], 'est__n_neighbors': [2,3,4,5,6]}, cv=10)
grid.fit(iris.iloc[:,:-1], iris.iloc[:,-1])
# grid.cv_results_
pd.DataFrame(grid.cv_results_).T
```

```python
from tensorflow.keras.datasets import mnist
```

```python
(X_train, y_train), (X_test, y_test) = mnist.load_data()
X_train = X_train.reshape(-1, 28*28)/255.0
X_test = X_test.reshape(-1, 28*28)/255.0
```

```python
from tensorflow.keras.models import Sequential, Model
from tensorflow.keras.layers import Dense
```

```python
model = Sequential()

layer1 = Dense(64, input_shape=(28*28,),activation='relu')
layer2 = Dense(64, activation = 'relu')
layer3 = Dense(64, activation = 'relu')
layer4 = Dense(64, activation = 'relu')
layer5 = Dense(10, activation = 'softmax')

model.add(layer1);
model.add(layer2);
model.add(layer3);
model.add(layer4);
model.add(layer5);

model.summary()

model.compile(loss='sparse_categorical_crossentropy', optimizer='adam')

model.fit(X_train, y_train, epochs=10)
```
