```python
import PIL
```

```python
from PIL import Image, ImageDraw, ImageFont
```

```python
fontTitle = ImageFont.truetype('AppleGothic.ttf', 250, encoding='utf-8')
fontContent = ImageFont.truetype('AppleGothic.ttf', 50, encoding='utf-8')
```

```python
width, height, margin = 800, 600, 10
```

```python
im = Image.open('calendar.png')
x = im.resize((width-margin*2, height-margin*2))
calendar = Image.new('RGB', (width, height), color='pink')
calendar.paste(x, (margin,margin))
```

```python
idc = ImageDraw.ImageDraw(calendar)
```

```python
delta = 111.5
```

```python
for j in range(6):
    for i in range(7):
        xy = ((10+i*delta, 10+j*delta), (10+(i+1)*delta, 10+delta))
        idc.rectangle(xy)
```

```python
idc.text((150,150), '달력', font=fontTitle, fill='yellow')

spac = 500
for i in range(7):
    idc.text((60 *spac*(i+1),50), '0', font=fontContent, fill='red')
```
