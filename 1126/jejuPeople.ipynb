```python
import pandas as pd
import matplotlib.pyplot as plt

plt.style.use('ggplot')
plt.rcParams['font.family'] = 'Noto Sans CJK JP'
```

```python
jejuP = pd.read_csv('./jejuPeople/jejuPeople.csv', encoding='cp949')
```

```python
a = jejuP.melt(jejuP.columns[:2], jejuP.columns[2:26], var_name='유동인구', value_name='유동인구 평균')
b = jejuP.melt(jejuP.columns[:2], jejuP.columns[26:], var_name='유동인구 비율', value_name='유동인구 비율 평균')
```

```python
a
```

```python
b
```

```python
c = b[['유동인구 비율', '유동인구 비율 평균']] # 팬시 인덱싱
```

```python
tidyJeju = pd.concat([a,c], axis=1) # a와 c를 concatnate
```

```python
tidyJeju
```

```python
tidyJeju[['유동인구', '유동인구 평균']].groupby('유동인구').sum().plot.bar(figsize=(40,20))
```
