```python
from sklearn.datasets import load_iris
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
data = load_iris()
```

```python
a = pd.DataFrame(data.data, columns=data['feature_names']) # convert nDarray & init columns from origin data's feature_names
```

```python
data['feature_names']
```

```python
b = pd.DataFrame(data['target'], columns=['target']) # 1dim convert DataFrame
```

```python
t = pd.concat([a,b], axis = 1)
```
